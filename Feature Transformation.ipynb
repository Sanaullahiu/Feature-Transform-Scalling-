# Import Liabiries

```python
import pandas as pd
import seaborn as sns
```

# Import Dataset

```python
df = pd.read_csv('supershops.csv')
df.head()
```

# Normalization imaplant with Dataset

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Min_max_scaler = MinMaxScaler(feature_range=(0, 1)) #range indicate which range i want
```

```python
#fit All Inter values with separately 


Min_scaler_fit = Min_max_scaler.fit(df[['Marketing Spend']])   #fit Marketing Values
Min_scaler_fit1 = Min_max_scaler.fit(df[['Administration']])    # fit Administration Values
Min_scaler_fit2 = Min_max_scaler.fit(df[['Transport']])         #fit Transport Values
Min_scaler_fit3 = Min_max_scaler.fit(df[['Profit']])            #Fit Profit Values

Min_scaler_fit
Min_scaler_fit1
Min_scaler_fit2
Min_scaler_fit3
```

```python
df.head()
```

```python
df['Marketing Spend'] = Min_max_scaler.transform(df[['Marketing Spend']]) # transform the data
df['Administration'] = Min_max_scaler.transform(df[['Administration']])
df['Transport'] = Min_max_scaler.transform(df[['Transport']])
df['Profit'] = Min_max_scaler.transform(df[['Profit']])
df.head()
```

# Standardrization imaplant with Dataset


```python
df1 = pd.read_csv('supershops.csv')
df1.head()
```

```python
# Import Standard Scaler from Sklearn
from sklearn.preprocessing import StandardScaler
Std_scaler= StandardScaler()
```

```python
Std_scale =Std_scaler.fit(df1[['Marketing Spend']])   #fit Marketing Values
Std_scale1 = Std_scaler.fit(df1[['Administration']])    # fit Administration Values
Std_scale2 = Std_scaler.fit(df1[['Transport']])         #fit Transport Values
Std_scale3 = Std_scaler.fit(df1[['Profit']]) 

Std_scale
Std_scale1
Std_scale2
Std_scale3
```

```python
df1.head()
```

```python
df1['Marketing Spend'] = Std_scaler.transform(df1[['Marketing Spend']]) # transform the data
df1['Administration'] = Std_scaler.transform(df1[['Administration']])
df1['Transport'] = Std_scaler.transform(df1[['Transport']])
df1['Profit'] = Std_scaler.transform(df1[['Profit']])
df1.head()
```
